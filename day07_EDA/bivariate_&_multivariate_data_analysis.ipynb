# 1- Scatterplot(Numerical-Numerical)

```python
import matplotlib.pyplot as plt
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
plt.title("Total Bill vs Tip")
plt.xlabel("total Bill")
plt.ylabel("tip")
plt.show()
```

# 2- Bar plot(Numerical-Categorical)

```python
titanic.head()
```

```python
sns.barplot(x=titanic['Pclass'],y = titanic['Age'])
```

```python
sns.barplot(data=titanic, x='Pclass', y='Age')
# Seaborn automatic grouping aur aggregation karega (mean by default)
# Reusable with hue, palette, etc.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 

sns.barplot(data=titanic, x='Pclass', y='Age', estimator=np.mean, hue='Pclass', palette='viridis', legend=True)


plt.title("Average Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Age")
plt.show()
# black rod showing confidence interval 
```

```python
sns.barplot(data=titanic, x='Pclass', y='Fare', estimator=np.mean, hue='Sex', palette='viridis', legend=True)


plt.title("Average Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```
